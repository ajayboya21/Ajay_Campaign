```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.value_counts()
```

```python
# 2. Univariate Analysis
# Survival distributionplt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution (0 = Did Not Survive, 1 = Survived)')
plt.savefig('survival_distribution.png')
plt.show()
```

```python
# Passenger class distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.savefig('pclass_distribution.png')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.savefig('age_distribution.png')
plt.show()
```

```python
# Fare distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.savefig('fare_distribution.png')
plt.show()
```

```python
# Boxplot for Age and Fare
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=df['Age'])
plt.title('Age Boxplot')
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Fare'])
plt.title('Fare Boxplot')
plt.savefig('boxplots.png')
plt.show()
```

```python
# 3. Bivariate Analysis
# Survival by Passenger Class
plt.figure(figsize=(8, 5))
sns.catplot(x='Pclass', hue='Survived', kind='count', data=df)
plt.title('Survival by Passenger Class')
plt.savefig('survival_by_pclass.png')
plt.show()
```

```python
#Survival by Sex
plt.figure(figsize=(6, 4))
sns.catplot(x='Sex', hue='Survived', kind='count', data=df)
plt.title('Survival by Sex')
plt.savefig('survival_by_sex.png')
plt.show()
```

```python
# Survival by Age
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title('Survival by Age')
plt.savefig('survival_by_age.png')
plt.show()
```

```python
# Survival by Fare
plt.figure(figsize=(10, 5))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival')
plt.savefig('fare_by_survival.png')
plt.show()
```

```python
# 4. Correlation Analysis
numerical_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.savefig('correlation_matrix.png')
plt.show()
```

```python
# 5. Pairplot for Numerical Variables
sns.pairplot(df[numerical_cols].dropna(), hue='Survived')
plt.savefig('pairplot.png')
plt.show()
```

```python
# 6. Additional Insights: Embarked and Family Size
# Survival by Embarked
plt.figure(figsize=(6, 4))
sns.catplot(x='Embarked', hue='Survived', kind='count', data=df)
plt.title('Survival by Embarkation Port')
plt.savefig('survival_by_embarked.png')
plt.show()
```

```python
# Create Family Size feature
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
plt.figure(figsize=(8, 5))
sns.countplot(x='FamilySize', hue='Survived', data=df)
plt.title('Survival by Family Size')
plt.savefig('survival_by_family_size.png')
plt.show()
```
